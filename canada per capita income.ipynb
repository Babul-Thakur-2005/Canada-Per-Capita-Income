```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("canada_per_capita_income.csv")
data.head()
```

# Check NULL values 

```python
data.isnull().sum()
```

# Create plot

```python
x = data['year']
y = data['per capita income (US$)']
m, c = np.polyfit(x, y, 1)
y_pred = m * x + c
plt.plot(x, y)
plt.plot(x, y_pred)
plt.xlabel("Year")
plt.ylabel("Income")
plt.title(" Canada per capita income (US$)")
plt.show()
```

```python
print(m, c)
```

# lets calculate income for 2005,2020,2022,2021,2025,2030?

```python
from sklearn import linear_model as lm
```

```python
reg = lm.LinearRegression()
reg.fit(data[['year']], data['per capita income (US$)'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
new = pd.DataFrame({'year' : [2005,2020,2022,2021,2025,2030]})
new['per capita income (US$)'] = reg.predict(new)
new
```
